# t-SNE vs PCA

Both t-SNE and PCA are dimensional reduction techniques that have different mechanisms and work best with different types of data.

## PCA (Principal Component Analysis):
- PCA is a linear technique that works best with data that has a linear structure.
- It seeks to identify the underlying principal components in the data by projecting onto lower dimensions.
- Minimizes variance and preserves large pairwise distances.
- PCA preserves the variance in the data, making it ideal for maintaining large pairwise distances.
- For more details, refer to the Principal Component Analysis (PCA) tutorial.

## t-SNE (t-Distributed Stochastic Neighbor Embedding):
- t-SNE is a nonlinear technique that focuses on preserving the pairwise similarities between data points in a lower-dimensional space.
- It is more concerned with preserving small pairwise distances.
- t-SNE is a good algorithm for visualizing complex high-dimensional data.
- Preserves the relationships between data points in the lower-dimensional space.

### Comparison:
- PCA preserves variance in the data.
- t-SNE preserves relationships between data points, making it better for visualization.

### How t-SNE Works:

1. High-Dimensional Similarity Calculation:
   - Models a point being selected as a neighbor of another point in higher dimensions.
   - Calculates pairwise similarity using a Gaussian kernel.
   - Points far apart have lower probabilities of being picked compared to close points.

2. Mapping to Lower Dimensions:
   - Maps high-dimensional data points to lower-dimensional space while preserving pairwise similarities.
   - Minimizes the divergence between the probability distributions of high-dimensional and low-dimensional spaces.

3. Optimization:
   - Uses gradient descent to minimize divergence.
   - Optimizes the lower-dimensional embedding to a stable state.
   - Creates clusters and sub-clusters of similar data points for visualization.

### Summary:
- t-SNE allows the visualization of high-dimensional data structures in a comprehensible way by clustering similar data points in the lower-dimensional space.


```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_swiss_roll
from sklearn.manifold import TSNE
```

```python
# Generate the Swiss Roll dataset
n_samples = 1500
X, color = make_swiss_roll(n_samples, noise=0.05)
```

```python
# Visualize the original 3D Swiss Roll
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')
sc = ax.scatter(X[:, 0], X[:, 1], X[:, 2], c=color, cmap=plt.cm.Spectral, s=20)
plt.title("Original 3D Swiss Roll Dataset", fontsize=16)
plt.colorbar(sc, label="Color Gradient")
plt.show()
```

```python
# Apply t-SNE for 2D dimensionality reduction
tsne = TSNE(n_components=2, random_state=42, perplexity=30, learning_rate=200, n_iter=1000)
X_tsne = tsne.fit_transform(X)
```

```python
# Visualize the 2D t-SNE result
plt.figure(figsize=(10, 8))
plt.scatter(X_tsne[:, 0], X_tsne[:, 1], c=color, cmap=plt.cm.Spectral, s=20)
plt.title("t-SNE Visualization of Swiss Roll Dataset (2D)", fontsize=16)
plt.xlabel("t-SNE Dimension 1", fontsize=12)
plt.ylabel("t-SNE Dimension 2", fontsize=12)
plt.colorbar(label="Color Gradient")
plt.grid(alpha=0.3)
plt.show()
```

# Steps in the Code

## Generate the Dataset:

We create the Swiss Roll dataset using make_swiss_roll from sklearn.datasets.
The dataset is a 3D spiral with a continuous color gradient for visualization.


## Visualize in 3D:

Use Matplotlib's 3D plotting capabilities to display the original high-dimensional dataset.


## Apply t-SNE:

Reduce the dimensionality to 2D while preserving pairwise relationships.
Parameters like perplexity=30 and learning_rate=200 are standard for such data.


## Visualize in 2D:

Display the 2D embedding produced by t-SNE. The unrolled structure of the Swiss Roll is revealed, showing how t-SNE successfully preserved the data relationships.
Key Insights




3D to 2D Mapping: t-SNE "unrolls" the Swiss Roll and clusters similar points together in 2D.
Gradient Preservation: The color gradient illustrates that points close in 3D remain close in the 2D representation.
